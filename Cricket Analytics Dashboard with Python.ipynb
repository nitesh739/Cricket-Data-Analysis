```python
# Season-wise Matches Played  

import matplotlib.pyplot as plt

season = df["season"].value_counts().sort_index()

plt.figure(figsize=(8,5))
plt.plot(season.index, season.values, marker='o')

plt.title("Season Wise Matches")
plt.xlabel("Season")
plt.ylabel("Matches")

plt.grid(True)

plt.show()
```

```python
# Team-wise Match Wins
import matplotlib.pyplot as plt

team_wins = df["winner"].value_counts()

plt.figure(figsize=(10,5))
plt.bar(team_wins.index, team_wins.values)

plt.title("Team Wise Wins",color='green')
plt.xlabel("Teams",color='black')
plt.ylabel("Number of Wins")

plt.xticks(rotation=45)

plt.grid(axis="y")

plt.show()
```

```python
#Top 10 Player of the Match    /Bar chart/
pom = df["player_of_match"].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(pom.index, pom.values)

plt.title("Top 10 Player of the Match")
plt.xlabel("Players")
plt.ylabel("Awards")

plt.xticks(rotation=45)

plt.grid(axis="y")

plt.show()
```

```python
#Total Score Distribution   /Histogram/

plt.figure(figsize=(9,5))

plt.hist(df["total_score"], bins=15)

plt.title("Total Score Distribution")
plt.xlabel("Total Score")
plt.ylabel("Frequency")

plt.grid()

plt.show()
```
