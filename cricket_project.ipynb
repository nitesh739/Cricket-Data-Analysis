```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\INDIA\OneDrive\Desktop\cricket_t20.xlsx")
print(df.head())
print(df.tail())
```

```python
#Data Cleaning
df.drop_duplicates(inplace=True)
df.isnull().sum()
```

```python
# total matches

total_matches = df['match_id'].nunique()
print(total_matches)
```

```python
#Total Runs

total_runs = df['runs_scored'].sum()
print(total_runs)
```

```python
#Total Wickets

total_wickets = df['wickets_taken'].sum()
print(total_wickets)
```

```python
#Total Venues

total_venues = df['venue'].nunique()
print(total_venues)

# top 10 vanues

df['venue'].value_counts().head(10)
```

```python
#Team Wise Runs

df.groupby('winner')['runs_scored'].sum()
```

```python
#Team Wise Matches Won
df.groupby('winner').size()
```

```python
df.groupby('venue').size()
```

```python
df.sort_values('runs_scored', ascending=False)
```

```python
# Filtering
df[df['season'] == 2023]
df[df['runs_scored'] < 123]
df[(df['season']==2023) & (df['runs_scored']<180)]
```
